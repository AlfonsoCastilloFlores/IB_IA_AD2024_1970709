```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Subir archivo manualmente
uploaded = files.upload()

# Obtener el nombre del archivo subido
ruta_imagen = list(uploaded.keys())[0]

# Cargar imagen
img = cv2.imread(ruta_imagen, cv2.IMREAD_COLOR)

# Convertir de BGR a RGB
imagen_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Separar canales RGB
b, g, r = cv2.split(imagen_rgb)

# Mostrar cada canal
plt.subplot(1, 3, 1)
plt.imshow(b, cmap='gray')
plt.title('Canal Azul')
plt.subplot(1, 3, 2)
plt.imshow(g, cmap='gray')
plt.title('Canal Verde')
plt.subplot(1, 3, 3)
plt.imshow(r, cmap='gray')
plt.title('Canal Rojo')
plt.show()

# Convertir la imagen RGB a escala de grises
imagen_gris = cv2.cvtColor(imagen_rgb, cv2.COLOR_RGB2GRAY)

# Mostrar ambas imágenes: original y en escala de grises
plt.subplot(1, 2, 1)
plt.imshow(imagen_rgb)
plt.title('Imagen RGB')
plt.subplot(1, 2, 2)
plt.imshow(imagen_gris, cmap="gray")
plt.title('Imagen en Escala de Grises')
plt.show()

# Segmentar una imagen con un umbral
umbral = 10  # Cambia este valor según el nivel deseado
_, imagen_segmentada = cv2.threshold(imagen_gris, umbral, 255, cv2.THRESH_BINARY)

# Mostrar imagen
plt.imshow(imagen_segmentada, cmap='gray')
plt.title('Imagen Segmentada')
plt.show()

```
